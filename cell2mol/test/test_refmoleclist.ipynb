```python
import cell2mol
import numpy as np
```

```python
ref_cell = np.load("BACZUB/Ref_Cell_BACZUB.cell", allow_pickle=True)
```

```python
ref_cell
```

```python
for ref in ref_cell.refmoleclist:
    print("===== REFERENCE MOLECULE =====")
    print(ref.formula, ref.type, ref.subtype)
    print(ref.labels)
    print(ref.parent_indices)
    if ref.iscomplex:
        print("===== METALS =====")
        for metal in ref.metals:
            print(metal.label, metal.type, metal.subtype, metal.parent_index)
            print("PARENT:\t:", metal.parent.formula, metal.parent.type, metal.parent.subtype)
        for ligand_idx, ligand in enumerate(ref.ligands):
            print(f"===== LIGAND {ligand_idx} =====")
            print(ligand.formula, ligand.type, ligand.subtype)
            print(ligand.labels)
            print(ligand.parent_indices)
            print("PARENT:\t:", ligand.parent.formula, ligand.parent.type, ligand.parent.subtype)
            for group_idx, group in enumerate(ligand.groups):
                print(f"===== GROUP {group_idx} =====")
                print(group.formula, group.type, group.subtype)
                print(group.labels)
                print(group.parent_indices)
                print("PARENT:\t:", group.parent.formula, group.parent.type, group.parent.subtype)
                print("===== ATOMS =====")
                for atom in group.atoms:
                    print(atom.label, atom.type, atom.subtype, atom.coord, atom.parent_index)
                    print("PARENT:\t:", atom.parent.formula, atom.parent.type, atom.parent.subtype)
    print("")
```
